```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score, accuracy_score,make_scorer
from sklearn.metrics import roc_curve
from sklearn.model_selection import GridSearchCV,ParameterGrid
from sklearn import tree
import math
```

```python
df = pd.read_csv("drive/MyDrive/Colab Notebooks/data_1.csv")
df
```

Distribution of the classes in the data set is:

```python
df['fetal_health'].value_counts()
```

As, shown above distribution is imbalanced, before splitting data is being shuffled as shown below:-

```python
df = df.sample(frac=1,random_state=0).reset_index(drop=True)
df
```

```python
#Dividing data frame as target and feature
target=df.columns[-1]
features=df.columns[:-1]
```

```python
#Splitting function which is used to split the data in 80:20 ratio of train data set and test data set
def splitting(features,target):
  splitting_index=int(0.8*df.shape[0])
  train_X=features[0:splitting_index]
  train_y=target[0:splitting_index]
  test_X=features[splitting_index:]
  test_y=target[splitting_index:]
  return train_X,train_y,test_X,test_y
```

```python
train_X,train_y,test_X,test_y=splitting(df[features],df[target])
```

A. Training

Que1

Classifying data set with decision tree classifier

```python
dtree=DecisionTreeClassifier(random_state=0)
dtree.fit(train_X,train_y)
```

Accuracy calculation

```python
dtree.score(test_X,test_y)
```

Rereiving predicted values from  test_X data set into y_predicted variable

```python
y_predict=dtree.predict(test_X)
```

Calculation of precision on trained decision tree for test_y and y_predict

```python
precision_value=precision_score(test_y,y_predict,average='macro')
precision_value
```

Calculation of recall on trained decision tree for test_y and y_predict

```python
recall_value=recall_score(test_y,y_predict,average='macro')
recall_value
```

```python
#Tree Size
dtree.tree_.node_count
```

Plotting of AUC-ROC Curve

```python
predict_values=dtree.predict_proba(test_X)
fpr={}
tpr={}
threshold={}
number_class=3
for x in range(number_class):
  fpr[x],tpr[x],threshold[x]=roc_curve(test_y,predict_values[:,x],pos_label=x+1)
  

```

```python
plt.plot(fpr[0], tpr[0],color='red', label='Class 1 vs Rest')
plt.plot(fpr[1], tpr[1], color='purple', label='Class 2 vs Rest')
plt.plot(fpr[2], tpr[2], color='green', label='Class 3 vs Rest')
plt.title('Multiclass ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend()
plt.plot()
```

Visualisation of decision tree and saving image as DT_A_1.jpg')

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (5,5), dpi=500)

tree.plot_tree(dtree,
               feature_names = train_X.columns, 
               class_names=np.unique(train_y).astype('str'),
               filled = True)
plt.savefig('DT_A_1.jpg')
```

Que2

```python
#Considering following depth for decision tree classification
depth=[1,4,6,9,10]

#Storing test accuracy and training accuracy in following two list for different depths
test_ac=[]
train_ac=[]

#Calculating accurace corresponding to different depths
for x in depth:
  trees=DecisionTreeClassifier(random_state=0,max_depth=x)
  trees.fit(train_X,train_y)
  
  test_ac.append(trees.score(test_X,test_y))
  train_ac.append(trees.score(train_X,train_y))
```

Plotting test accuracy , training accuracy vs depth graph

```python
plt.plot(depth,test_ac,label='Test Accuracy',color='red')
plt.plot(depth,train_ac,label='Training Accuracy',color='purple')
plt.xlabel('Tree Maximum Depths')
plt.ylabel('Accuracies')
plt.legend()
plt.show()
```

B. **Post** **Pruning**

Q2. Post Pruning and Pre Pruning Technique is being used

Post-Pruning Tree

```python
grid_search_ccp=GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=0,max_depth=16,max_features='log2',min_samples_split=8),
    scoring=make_scorer(accuracy_score),
    param_grid=ParameterGrid({"ccp_alpha":[
                                     [ccp_alpha] for ccp_alpha in np.arange(0.0001,0.001,0.0001)
    ]}),
    
)
```

```python
grid_search_ccp.fit(train_X,train_y)
```

```python
grid_search_ccp.best_params_
```

```python
yccp_predict=grid_search_ccp.predict(test_X)
```

```python
#Precision
precision_value=precision_score(test_y,yccp_predict,average='macro')
precision_value
```

```python
#Recall
recall_value=recall_score(test_y,yccp_predict,average='macro')
recall_value
```

```python
#Accuracy
grid_search_ccp.score(test_X,test_y)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (3,3), dpi=300)

tree.plot_tree(grid_search_ccp.best_estimator_,
               feature_names = train_X.columns, 
               class_names=np.unique(train_y).astype('str'),
               filled = True)
plt.savefig('DT_B_2_CC.jpg')
```

**Pre Pruning**

```python
grid_search_min_samplesplit=GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=0,max_depth=16,max_features='log2',min_samples_split=8),
    scoring=make_scorer(accuracy_score),
    param_grid=ParameterGrid({"min_samples_split":[
                                     [min_samples_split] for min_samples_split in np.arange(2,20,2)
    ]}),
    
)
```

```python
grid_search_min_samplesplit.fit(train_X,train_y)
```

```python
grid_search_min_samplesplit.best_params_
```

```python
yminsample_predict=grid_search_min_samplesplit.predict(test_X)
```

```python
#Precision
precision_value=precision_score(test_y,yminsample_predict,average='macro')
precision_value
```

```python
#Recall
recall_value=recall_score(test_y,yminsample_predict,average='macro')
recall_value
```

```python
#Accuracy
grid_search_min_samplesplit.score(test_X,test_y)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (3,3), dpi=300)

tree.plot_tree(grid_search_min_samplesplit.best_estimator_,
               feature_names = train_X.columns, 
               class_names=np.unique(train_y).astype('str'),
               filled = True)
plt.savefig('DT_B_2_min_samples_split.jpg')
```
